- Initiate a for loop over the list of dataset names. Inside the loop:

```python
# Write your own code
csv_names = ['data_quiz/wdi_agedist_1980.csv', 'data_quiz/wdi_agedist_2000.csv', 'data_quiz/wdi_agedist_2020.csv']
list_newnames = ['data_excel/wdi_agedist_1980.xlsx', 'data_excel/wdi_agedist_2000.xlsx', 'data_excel/wdi_agedist_2020.xlsx']

index = 0
for index in range(len(csv_names)):
    dataset = pd.read_csv(csv_names[index])
    filename = list_newnames[index]
    dataset.to_excel(filename)
    index = index + 1
```

(f) Your goal is to open each dataset, create a new variable, then plot


<font size = "3">

- Create a list with the ".csv" dataset names (including their relative subfolder "data_quiz/)
- Start an index

$\qquad$ index = 0

- Initiate a for loop over the list of dataset names. Inside the loop:
    - Read the dataset and store it in an object called "dataset"
    - Create two new variables as
    
    $\qquad$ percent_ages0to14 = dataset["percent_ages0to14"] <br>

    - Calculate the proportion of the population 15 or older by computing "100 -  percent_ages0to14".
    - Plot a histogram and label the axes
    - Use plt.show() to display the plot
    - Write

    $\qquad$ index = index + 1


This should produce a histogram for each dataset

```python
# Write your own code
csv_names = ['data_quiz/wdi_agedist_1980.csv', 'data_quiz/wdi_agedist_2000.csv', 'data_quiz/wdi_agedist_2020.csv']
dataset_years = [1980, 2000, 2020]

index = 0
for index in range(len(csv_names)):
    dataset = pd.read_csv(csv_names[index])
    percent_ages0to14 = dataset["percent_ages0to14"]
    proportion15orolder = 100 -  percent_ages0to14
    plt.hist(proportion15orolder)
    plt.xlabel("Proportion of people older than 15")
    plt.ylabel("Frequency")
    plt.title(dataset_years[index])
    plt.show()
    index = index + 1
```
